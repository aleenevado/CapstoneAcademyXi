```python
import pandas as pd

df = pd.read_csv('Drug_clean.csv')

df.head()
```

```python
# Step 1: Exploring the data

info = df.info()
desc_stats = df.describe()
missing_values = df.isnull().sum()
full_duplicates = df[df.duplicated(keep=False)]

info, desc_stats, missing_values, full_duplicates
```

```python
unique_counts = df.nunique()
print(unique_counts)

# Extracting unique values for 'Type', 'Form', and 'Indication' columns
unique_types = df['Type'].unique()
unique_forms = df['Form'].unique()
unique_indications = df['Indication'].unique()

unique_types, unique_forms, unique_indications
```

```python
# Dropping rows where 'Type' or 'Indication' has '\r\r\n'
df = df[(df['Type'] != '\r\r\n') & (df['Indication'] != '\r\r\n')]
df.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Plotting visualizations to understand the relationships

# Creating a figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Plotting the relationship of each independent variable with the dependent variable 'Price'
variables = ['EaseOfUse', 'Effective', 'Reviews', 'Satisfaction']
for var, ax in zip(variables, axes.flatten()):
    # Scatter plot for each variable vs. Price
    sns.scatterplot(x=df[var], y=df['Price'], ax=ax, alpha=0.6)
    
    # Plotting the regression line
    sns.regplot(x=df[var], y=df['Price'], ax=ax, scatter=False, color='red')
    
    # Setting the title for each subplot
    ax.set_title(f'Relationship between {var} and Price')
    ax.set_ylabel('Price')

plt.tight_layout()
plt.show()
```

```python
df.hist(bins=10, figsize=(15,10))
plt.show()
```

```python
# Step 2: Removing outliers from the 'Price' column
# A common method to define outliers is to look for values outside 1.5 * IQR (Interquartile Range) above Q3 and below Q1

Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

# Defining boundaries for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtering the dataset to remove outliers
df = df[(df['Price'] >= lower_bound) & (df['Price'] <= upper_bound)]

# Visualizing the new distribution of 'Price' without outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Price'])
plt.title('Boxplot of Drug Prices without Outliers')
plt.xlabel('Price')
plt.show()
df.shape
```

```python
import statsmodels.api as sm

# Selecting the independent variables for the model
# For simplicity, we're initially considering only the numerical variables as predictors
X = df[['EaseOfUse', 'Effective', 'Reviews', 'Satisfaction']]
y = df['Price']

# Adding a constant to the model (intercept)
X = sm.add_constant(X)

# Building the Ordinary Least Squares (OLS) model
model = sm.OLS(y, X).fit()

# Summary of the regression model
model_summary = model.summary()

# Extracting the Adjusted R-squared value
adjusted_r_squared = model.rsquared_adj

model_summary, adjusted_r_squared
```

```python
# Creating a QQ plot to check the normality of residuals
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

sm.qqplot(model.resid, line='s', ax=ax[0])
ax[0].set_title('QQ Plot of Residuals')

# Residuals vs Fitted values plot to check homoscedasticity
sns.residplot(x=model.fittedvalues, y=model.resid, lowess=True, ax=ax[1], line_kws={'color': 'red', 'lw': 1})
ax[1].set_title('Residuals vs Fitted')
ax[1].set_xlabel('Fitted values')
ax[1].set_ylabel('Residuals')

# Plotting the homoscedasticity trend line
ax[1].axhline(y=0, color='grey', linestyle='dashed')

# Show the plots
plt.tight_layout()
plt.show()
```

The summary shows the results of an OLS regression analysis after removing outliers only from the 'Price' variable.

1. **Model Fit**:
   - The `R-squared` value is 0.012, which means that only 1.2% of the variability in 'Price' is explained by the model. This is a very low value, indicating a poor fit.
   - The `Adj. R-squared` is 0.006, adjusting for the number of predictors. This value being close to the R-squared suggests there are not too many predictors inflating the fit.

2. **Coefficients**:
   - `const` (Intercept): The model predicts a 'Price' of approximately 72.0496 when all other variables are zero, and this value is statistically significant (p < 0.05).
   - Among the predictors, only 'Effective' has a p-value (0.015) below the typical alpha level of 0.05, suggesting it has a statistically significant relationship with 'Price'.
   - The coefficients for 'EaseOfUse', 'Reviews', and 'Satisfaction' are not statistically significant given their p-values are greater than 0.05.

3. **Model Significance**:
   - The `F-statistic` is 1.925 with a p-value of 0.105, indicating that the model is not statistically significant as a whole at the conventional 0.05 alpha level.

4. **Residual Tests**:
   - The `Omnibus` test has a very low p-value, suggesting the residuals are not normally distributed.
   - `Skew` (1.382) and `Kurtosis` (4.358) also suggest that the residuals are not normally distributed.
   - The `Durbin-Watson` statistic (1.429) is close to 2, which implies there is no strong autocorrelation in the residuals.
   - The `Jarque-Bera` test also indicates that the residuals are not normally distributed (p-value is very low).

5. **Multicollinearity**:
   - The `Condition Number` is quite large, indicating potential issues with multicollinearity.

In summary, while removing outliers from 'Price' may have cleaned up some extreme values, the model's ability to explain the variability in 'Price' is still very limited. The lack of normality in residuals and the potential multicollinearity are concerning. It suggests that further work is needed, such as transforming variables, adding other relevant predictors, or considering non-linear models.

Observing the regression results and the Q-Q plots provided, here are some key points:

### OLS Regression Results Observations:

- **R-squared is very low**: The R-squared value is 0.004, which means the model explains only 0.4% of the variance in the dependent variable. This suggests that the model does not have a good fit.
  
- **Adjusted R-squared is negative**: This can happen when the model is overly complex relative to the size of the sample and/or the model does not explain the variation in the dependent variable.
  
- **Significance of coefficients**: The p-values for the coefficients of 'EaseOfUse' and 'Effective' are greater than 0.05, indicating that these predictors are not statistically significant at the 5% significance level. However, the 'const' (intercept) is significant.
  
- **Large standard errors**: Some predictors have large standard errors relative to the coefficients, which suggests there might be high uncertainty in the estimation of these coefficients.
  
- **Condition number**: The condition number is large (1.41e+03), which may indicate that there is multicollinearity or other numerical problems. Multicollinearity occurs when predictor variables are highly correlated with each other.

### Q-Q Plot Observations:

- **First Q-Q Plot**: The points in the first Q-Q plot follow the theoretical line at the center of the distribution but deviate at the tails. This indicates that the residuals have heavier tails than a normal distribution (possible presence of outliers or extreme values).

- **Second Q-Q Plot**: Similar to the first, this plot shows that the residuals largely follow the theoretical line, but with some deviation, especially in the upper tail. This suggests that the residuals may not be perfectly normally distributed.

- **Third Q-Q Plot**: This plot shows a clear deviation from the theoretical line across the distribution, particularly in the tails. This is an indication that the residuals are not normally distributed and that the data may have non-linearities or be influenced by outliers.

### General Observations:

- The model might benefit from further investigation and possibly the inclusion of additional relevant predictors, transformation of variables, or the use of alternative modeling techniques that are more robust to non-normal distributions.
  
- The presence of non-normally distributed residuals suggests that some assumptions of OLS may not be fully met, which can affect the validity of hypothesis tests and confidence intervals.

- The interpretation of the regression and Q-Q plots should be done in the context of the domain from which the data originates. Domain knowledge could provide insights into why certain variables are not significant or why the residuals are distributed as they are.

The overall assessment would suggest revisiting the model specification, considering data transformation, outlier removal, or non-linear modeling techniques to improve model fit and adherence to OLS assumptions.
