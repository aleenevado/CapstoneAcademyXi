```python
import pandas as pd

df = pd.read_csv('Drug_clean.csv')

df.head()
```

```python
# Step 1: Exploring the data

# Basic information about the dataset
info = df.info()

# Descriptive statistics for numerical columns
desc_stats = df.describe()

# Checking for missing values
missing_values = df.isnull().sum()

# Displaying the results
info, desc_stats, missing_values
```

```python
# Step 2: Removing outliers from the 'Price' column
# A common method to define outliers is to look for values outside 1.5 * IQR (Interquartile Range) above Q3 and below Q1
import matplotlib.pyplot as plt
import seaborn as sns

Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

# Defining boundaries for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtering the dataset to remove outliers
df_no_outliers = df[(df['Price'] >= lower_bound) & (df['Price'] <= upper_bound)]

# Visualizing the new distribution of 'Price' without outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_no_outliers['Price'])
plt.title('Boxplot of Drug Prices without Outliers')
plt.xlabel('Price')
plt.show()

# Displaying the shape of the dataset before and after outlier removal
original_shape = df.shape
new_shape = df_no_outliers.shape

(original_shape, new_shape)
```

```python
import statsmodels.api as sm

# Selecting the independent variables for the model
# For simplicity, we're initially considering only the numerical variables as predictors
X = df_no_outliers[['EaseOfUse', 'Effective', 'Reviews', 'Satisfaction']]
y = df_no_outliers['Price']

# Adding a constant to the model (intercept)
X = sm.add_constant(X)

# Building the Ordinary Least Squares (OLS) model
model = sm.OLS(y, X).fit()

# Summary of the regression model
model_summary = model.summary()

# Extracting the Adjusted R-squared value
adjusted_r_squared = model.rsquared_adj

model_summary, adjusted_r_squared
```

```python
# Plotting visualizations to understand the relationships and the model better

# Creating a figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Plotting the relationship of each independent variable with the dependent variable 'Price'
variables = ['EaseOfUse', 'Effective', 'Reviews', 'Satisfaction']
for var, ax in zip(variables, axes.flatten()):
    # Scatter plot for each variable vs. Price
    sns.scatterplot(x=df_no_outliers[var], y=df_no_outliers['Price'], ax=ax, alpha=0.6)
    
    # Plotting the regression line
    sns.regplot(x=df_no_outliers[var], y=df_no_outliers['Price'], ax=ax, scatter=False, color='red')
    
    # Setting the title for each subplot
    ax.set_title(f'Relationship between {var} and Price')
    ax.set_ylabel('Price')

# Adjust layout to prevent overlap
plt.tight_layout()

# Showing the plots
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the dataset without outliers
file_path = 'Drug_clean.csv'
df_no_outliers = pd.read_csv(file_path)

# Considering the same independent variables as before for the model
X = df_no_outliers[['EaseOfUse', 'Effective', 'Reviews', 'Satisfaction']]
y = df_no_outliers['Price']

# Adding a constant to the model (intercept)
X = sm.add_constant(X)

# Building the Ordinary Least Squares (OLS) model
model = sm.OLS(y, X).fit()

# Creating a QQ plot to check the normality of residuals
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

sm.qqplot(model.resid, line='s', ax=ax[0])
ax[0].set_title('QQ Plot of Residuals')

# Residuals vs Fitted values plot to check homoscedasticity
sns.residplot(x=model.fittedvalues, y=model.resid, lowess=True, ax=ax[1], line_kws={'color': 'red', 'lw': 1})
ax[1].set_title('Residuals vs Fitted')
ax[1].set_xlabel('Fitted values')
ax[1].set_ylabel('Residuals')

# Plotting the homoscedasticity trend line
ax[1].axhline(y=0, color='grey', linestyle='dashed')

# Show the plots
plt.tight_layout()
plt.show()

# Checking the mean of the residuals
mean_residuals = model.resid.mean()

mean_residuals
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Initializing encoders
one_hot_encoder = OneHotEncoder()
label_encoder = LabelEncoder()

# One-hot encoding for 'Form', 'Indication', and 'Type'
one_hot_encoded = pd.get_dummies(df_no_outliers[['Form', 'Indication', 'Type']], drop_first=True)

# Label encoding for 'Drug' and 'Condition'
df_no_outliers['Drug_encoded'] = label_encoder.fit_transform(df_no_outliers['Drug'])
df_no_outliers['Condition_encoded'] = label_encoder.fit_transform(df_no_outliers['Condition'])

# Combining the encoded features with the original dataframe
encoded_data = pd.concat([df_no_outliers, one_hot_encoded], axis=1)

# Dropping the original categorical columns
encoded_data.drop(['Form', 'Indication', 'Type', 'Drug', 'Condition'], axis=1, inplace=True)

# Displaying the new dataframe
encoded_data.head()
```

```python
# Setting the plot style
sns.set(style="whitegrid")

# Calculating the correlation matrix
corr_matrix = encoded_data.corr()

# Plotting the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
#'encoded_data' original DataFrame
# Select all columns except 'Price' by using the column names
columns_to_use = [col for col in encoded_data.columns if col != 'Price']
data_pred = encoded_data[columns_to_use]
data_pred.head()
```
