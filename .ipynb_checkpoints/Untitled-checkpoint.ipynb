```python
import pandas as pd

data = pd.read_csv('Drug_clean.csv')
data.head()
```

```python
data.info()
```

```python
# Checking for missing values
missing_values = data.isnull().sum()

# Checking data types for encoding categorical variables
data_types = data.dtypes

missing_values, data_types
```

```python
print(data[['Form','Indication','Type']].describe())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(24,5))

for xcol, ax in zip(['EaseOfUse', 'Effective', 'Reviews', 'Satisfaction'], axes):
    data.plot(kind='scatter', x=xcol, y='Price', ax=ax, alpha=0.4, color='b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12,3))

for xcol, ax in zip([ 'Form', 'Indication', 'Type'], axes):
    data.plot(kind='scatter', x=xcol, y='Price', ax=ax, alpha=0.4, color='b')
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Initializing encoders
one_hot_encoder = OneHotEncoder()
label_encoder = LabelEncoder()

# One-hot encoding for 'Form', 'Indication', and 'Type'
one_hot_encoded = pd.get_dummies(data[['Form', 'Indication', 'Type']], drop_first=True)

# Label encoding for 'Drug' and 'Condition'
data['Drug_encoded'] = label_encoder.fit_transform(data['Drug'])
data['Condition_encoded'] = label_encoder.fit_transform(data['Condition'])

# Combining the encoded features with the original dataframe
encoded_data = pd.concat([data, one_hot_encoded], axis=1)

# Dropping the original categorical columns
encoded_data.drop(['Form', 'Indication', 'Type', 'Drug', 'Condition'], axis=1, inplace=True)

# Displaying the new dataframe
encoded_data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the plot style
sns.set(style="whitegrid")

# Calculating the correlation matrix
corr_matrix = encoded_data.corr()

# Plotting the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
predictors_for_plots = ['EaseOfUse', 'Effective', 'Reviews', 'Satisfaction']

plt.figure(figsize=(20, 10))
for i, predictor in enumerate(predictors_for_plots, 1):
    plt.subplot(2, 2, i)
    sns.scatterplot(data=encoded_data, x=predictor, y='Price')
    plt.title(f'Price vs {predictor}')

plt.tight_layout()
plt.show()
```
