```python
# It appears that the session was reset, which means we lost the previous state including all imports and variables.
# Let's start by re-importing the necessary libraries and re-loading the dataset.

import pandas as pd
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the dataset without outliers
file_path = 'Drug_clean.csv'
df_no_outliers = pd.read_csv(file_path)

# Considering the same independent variables as before for the model
X = df_no_outliers[['EaseOfUse', 'Effective', 'Reviews', 'Satisfaction']]
y = df_no_outliers['Price']

# Adding a constant to the model (intercept)
X = sm.add_constant(X)

# Building the Ordinary Least Squares (OLS) model
model = sm.OLS(y, X).fit()

# Creating a QQ plot to check the normality of residuals
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

sm.qqplot(model.resid, line='s', ax=ax[0])
ax[0].set_title('QQ Plot of Residuals')

# Residuals vs Fitted values plot to check homoscedasticity
sns.residplot(x=model.fittedvalues, y=model.resid, lowess=True, ax=ax[1], line_kws={'color': 'red', 'lw': 1})
ax[1].set_title('Residuals vs Fitted')
ax[1].set_xlabel('Fitted values')
ax[1].set_ylabel('Residuals')

# Plotting the homoscedasticity trend line
ax[1].axhline(y=0, color='grey', linestyle='dashed')

# Show the plots
plt.tight_layout()
plt.show()

# Checking the mean of the residuals
mean_residuals = model.resid.mean()

mean_residuals
```
